```python
x1= df.YearsExperience
```

```python
y= df.Salary
y.head()
```

#### y= mx+c can be thought of as y= mx1 + x0*c where x0 is always 1

```python
X= sm.add_constant(x1)
```

```python
X.head()
```

```python
model=sm.OLS(y,X).fit()
```

```python
model.summary()
```

```python
x1.head()
```

```python
# OLS line
#y=mx+c
y_pred= 9449.9623*x1 + 2.579e+04
```

```python
9449.9623*500 + 2.579e+04
```

```python
np.mean(y)
```

```python
plt.scatter(x1,y,label='Original data');
plt.plot(x1,y_pred,label='Line of best fit',c='r')
plt.legend()
```

```python
plt.scatter(x1,y,label='Original Data')
plt.plot(x1,y_pred,c='r',label='Reg. Line')
#plt.axhline(np.mean(y),c='g',label='Mean')
plt.legend();
```

```python
plt.scatter(x1,y,label= 'Original Data dist');
plt.plot(x1,y_pred,c='r',label='Regression Line')
#plt.axhline(np.mean(y),c='g',label='Mean Line')

plt.legend();
```

```python
plt.scatter(x1,y,label= 'Data')
plt.plot(x1,y_pred,c='r',label='My Regression Line')
plt.xlabel("Years of Experience",c='darkblue')
plt.ylabel('Salary',c='darkblue')
plt.title('Salary v Years of Experience',c='darkblue')
plt.legend();
```
